```python
papers = pd.DataFrame(papers_soup['Papers'])
```

```python
papers
```

```python
papers.count()
```

```python
papers.groupby('title').count()
```

```python
papers = papers[~pd.isnull(papers.soup)].drop('title', axis=1)
```

```python
papers
```

```python
papers['soup'] = papers.soup.map(lambda x: BeautifulSoup(x))
```

```python
papers['sections'] = papers.soup.map(lambda x: scipdf.parse_sections(x))
papers['abstract'] = papers.soup.map(lambda x: scipdf.parse_abstract(x) if x is not None else '')
```
